```python
import numpy as np
import scipy.special as sp
import matplotlib.pyplot as plt
import random
import operator
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mu=0.0
std = 0.1
def gaussian_noise(x,mu,std):
    noise = np.random.normal(mu, std, size = x.shape)
    x_noisy = x + noise
    return x_noisy 
```

```python
plt.rcParams.update(plt.rcParamsDefault)
n = np.linspace(0, 8, 100)
array1 = gaussian_noise(n,mu,std)
plt.plot(n, sp.jv(0, n))
plt.plot(n, sp.jv(0,array1))
plt.xlim((0, 8))
plt.ylim((-0.5, 1.1))
plt.legend(('${J}_0(x)$','${J}_0(x) with little noise$' ),loc = 1)
plt.xlabel('$n$')
plt.ylabel('${J}_n(x)$')
plt.grid(True)
plt.show()
```

### 10 data

```python
randarray = np.random.choice(array1, 10, replace=False)
df = pd.DataFrame({"y" : randarray, "x" : np.arange(0, len(randarray))})
df_val = df.sample(n = int(len(randarray)*2/10)).sort_values(by='x')
df_train = df[~df.isin(df_val)].dropna()
x_train = df_train.x.values
x_train = x_train.reshape(-1,1)
y_train = df_train.y.values

x_val = df_val.x.values
x_val = x_val.reshape(-1,1)
y_val = df_val.y.values
polynomial_features= PolynomialFeatures(degree=8)
x_poly_train = polynomial_features.fit_transform(x_train)
polynomial_features.fit(x_poly_train,y_train)
linreg = LinearRegression()
linreg.fit(x_poly_train,y_train)
y_pred = linreg.predict(x_poly_train)
plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,y_pred,color='blue')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(y_train,y_pred))
r2 = r2_score(y_train,y_pred)
print('Rmse is: ' + str(rmse))
print('R2 score is: ' +str(r2))
```

```python
x_poly_val = polynomial_features.fit_transform(x_val)
polynomial_features.fit(x_poly_val,y_val)
y_pred = linreg.predict(x_poly_val)
plt.scatter(x_val,y_val,color='red')
plt.plot(x_val,y_pred,color='blue')
plt.show()
```

```python
rmse = np.sqrt(mean_squared_error(y_val,y_pred))
r2 = r2_score(y_val,y_pred)
print('Rmse is: ' + str(rmse))
print('R2 score is: ' +str(r2))
```
